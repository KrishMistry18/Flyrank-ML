**Lane:** Structured Content Archetype Clustering (Lane 3)

I'm going with clustering because I want to look for hidden patterns across the content inventory using unsupervised learning. Instead of just trying to predict one specific thing like a traffic drop, I want to see if the data naturally groups into recurring "archetypes" (like 'stale but highly visible', 'hidden gems', or 'thin content with high engagement'). I'm curious what kind of shape the content actually takes when you let the metrics group themselves.

## 2. The question: decision, action, cost of a wrong call

**The Research Question:** What distinct performance and structural archetypes exist across our content inventory?

**Unit of Analysis:** A single page (pseudonymized content item) over a 90-day window.

**The Decision & Action:** 
This helps content editors and SEO managers decide *how* to treat different chunks of their inventory at scale. If a page falls into a specific archetype (e.g. 'high impressions, low CTR, thin content'), the action someone takes is mapped directly to that cluster—like rewriting the title, merging it with another page, or just leaving it alone to protect its traffic. 

**Cost of a wrong call:**
If we cluster badly or mislabel an archetype, an editor might waste hours rewriting a page that was actually doing fine, or worse, they might accidentally prune or mess with a "champion" page and tank its existing traffic. 

**Why ML helps here:**
A simple rule (like `if word_count < 500 and clicks < 10 then prune`) is too rigid and misses the nuance of how different metrics interact. By using clustering (like K-Means), we can look at a bunch of signals at once—age, engagement, search volume, word count—and let the algorithm find the natural dividing lines that a human writing `if/else` statements would miss.
